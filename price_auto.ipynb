# Описание проекта:

**Цель проекта:**
разработать модель предсказания стоимости автомобиля на вторичном рынке.

**Данные**

`train.csv` - информация о характеристиках автомобилей (-440000), которые будут использоваться в качестве обучающих данных. </br>
`test.csv` - информация о характеристиках автомобилей (~110000), которые будут использоваться в качестве тестовых данных. </br>
Ваша задача - предсказать значение 
`sellingprice` для каждого автомобиля из этого датасета.</br>
`sample_submission.csv` - пример файла предсказаний в правильном формате.</br>
`vin` - идентификатор автомобиля в тестовом наборе.</br>
`sellingprice` - Целевой признак. Для каждого автомобиля предскажите числовое значение стоимости продажи.

**Описание полей данных**

`Date`: Год выпуска автомобиля.</br>
`Make`: Марка автомобиля.</br>
`Model`: Модель автомобиля определенной.</br>
`Trim`: Версия модели.</br>
`Body`: Тип кузова транспортного средства относится к форме и модели конкретной марки автомобиля.</br>
`Transmission`: Коробка передач.</br>
`VIN`: Идентификационный номер транспортного средства.</br>
`State`: Название штата.</br>
`Condition`: Состояние автомобиля на момент аукциона.</br>
`Odometer`: Расстояние, пройденное автомобилем с момента выпуска.</br>
`Color`: Цвет кузова автомобиля.</br>
`Interior`: Цвет салона автомобиля.</br>
`Seller`: Продавец автомобиля, автосалоны.</br>
`sellingprice`: Цена, по которой автомобиль был продан на аукционе.</br>
`saledate`: Дата продажи автомобиля.

**Метрика качества**

MAPE (Mean Absolute Percentage Error)

##### Рассмотим данные  `year` (год выпуска автомобиля):

```python
# создание нового признака "количество лет":
def year_count(df):
    
    shape = df.shape[0]
    
    # выделим год продажи автомобиля и найдем его возраст  
    # на момент продажи
    df['year_sale'] = df['saledate'].dt.year
    df['year_count'] = df['year_sale'] - df['year']
    
    print('Уникальные значения:',df['year_count'].unique())
    
    # количество, где год = -1:
    print('Количество значений, где год = -1:', 
          df.query('year_count == -1')['year'].count())
    
    # замена значений:
    df['year_count'] = df['year_count'].replace(-1, 0)
    print('Проверка: количество значений, где год = -1:', 
          df.query('year_count == -1')['year'].count())
    
year_count(df)
```

```python
# функция заполнения пропущенных значений 'odometer' медианой по году-марке-модели:
def fillna_with_median(df, column_name, groupby_columns):
    
    print('Количество пропусков было:', df[column_name].isna().sum())
    
    # создадим сводную таблицу с медианными значениями:
    pivot_table = df.pivot_table(index=groupby_columns, values=column_name, aggfunc='median')

    # объединим сводную таблицу с исходной таблицей по ключевым полям:
    df = df.merge(pivot_table, how='left', left_on=groupby_columns, 
                  right_index=True, suffixes=('', '_median'))

    # заменим пропущенные значения - медианой:
    outlier_condition = (df[column_name].isna()) 

    df.loc[outlier_condition, column_name] = df.loc[outlier_condition, column_name + '_median']
    
    # удалим созданную колонку с медианными значениями:
    df.drop([column_name + '_median'], axis=1, inplace=True)
    
    print('Количество пропусков стало:', df[column_name].isna().sum())
    
    return df

# заполнение пропущенных значений медианой: 
#df = fillna_with_median(df, 'odometer', ['make', 'model', 'year'])
```

```python
# функция заполнения выбивающихся значений 'odometer' медианой:
def replace_outliers_with_median(df, column_name, groupby_columns):
    
    print('Количество машин с пробегом (odometer) больше 425000  было:',
          df[df[column_name] > 425000]['year'].count())
    
    # создадим сводную таблицу с медианными значениями:
    pivot_table = df.pivot_table(index=groupby_columns, values=column_name, aggfunc='median')

    # объединим сводную таблицу с исходной таблицей по ключевым полям:
    df = df.merge(pivot_table, how='left', left_on=groupby_columns, 
                  right_index=True, suffixes=('', '_median'))

    # заменим выбросы медианными значениями:
    outlier_condition = df[column_name] > 400000

    df.loc[outlier_condition, column_name] = df.loc[outlier_condition, column_name + '_median']

    # удалим временные столбцы:
    df.drop([column_name + '_median'], axis=1, inplace=True)
    
    print('Количество машин с пробегом (odometer) больше 425000  стало:',
          df[df[column_name] > 425000]['year'].count())
    
    return df

df_test = replace_outliers_with_median(df_test, 'odometer', ['make', 'model', 'year'])
```
